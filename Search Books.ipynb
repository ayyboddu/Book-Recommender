```python
#takes the line and only returns the fields we care about
def parse_fields(line):
    data = json.loads(line)
    return {
        "book_id":data["book_id"],
        "title":data["title_without_series"],
        "ratings":data["ratings_count"],
        "url":data["url"],
        "cover_image":data["image_url"]
    }
```

```python
#go line by line and parse each line

books_titles = []
with open("goodreads_books.json") as f:
    while True:
        line = f.readline()
        if not line:
            break
        fields = parse_fields(line)
        
        try:
            ratings = int(fields["ratings"])
        except ValueError:
            continue
        if ratings > 15:
            books_titles.append(fields) #only takes books with more than 15 ratings
```

```python
import pandas as pd

titles = pd.DataFrame.from_dict(books_titles) #book titles will be a list of dictionaries. this will turn each dict into a line
```

```python
titles["ratings"]=pd.to_numeric(titles["ratings"]) #turn ratings into a numeric column so we can run comparisons
```

```python
#modify titles to get rid of characters that doesn't fall in the a-z A-Z 0-9 set
titles["mod_title"] = titles["title"].str.replace("[^a-zA-Z0-9 ]","",regex=True)
```

```python
titles
```

```python
#make everyhitng lower case
titles["mod_title"] = titles["mod_title"].str.lower()
```

```python
#replace anything with multiple spaces in a row with one space
titles["mod_title"] = titles["mod_title"].str.replace("\s+"," ", regex=True)
```

```python
titles
```

```python
#only take titles whos title length is >0
titles = titles[titles["mod_title"].str.len() > 0]
```

```python
#dump to json to use in future sessions
titles.to_json("books_titles.json")
```

# Create Search Engine

```python
#search up tf idf - it's a statistical measure that evaluates the relevency of a word to the search

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(titles["mod_title"])
```

```python
#turn search query into a vector and do a comparison of vector

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re

def make_clickable(val):
    return '<a target = "_blank" href="{}">Goodreads</a>'.format(val)

def show_image(val):
    return '<img src="{}" width=50></img>'.format(val)


def search(query,vectorizer):
    processed = re.sub("[^a-zA-z0-9 ]", "", query.lower())
    query_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(query_vec,tfidf).flatten()
    # finding the 10 largest similarities.
    indices = np.argpartition(similarity,-10)[-10:]
    results = titles.iloc[indices]
    #there's a lot of duplicates so to solve this we take the one with the highest number of ratings
    results = results.sort_values("ratings", ascending = False)
    return results.head(5).style.format({'url':make_clickable, 'cover_image':show_image})
```

```python
search("east of eaden", vectorizer)
```
